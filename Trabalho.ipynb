## Total de medalhas por Atletas brasileiros
### Vamos trabalha sÃ³ com as colunas Atleta e Medalha
#### Aqui vamos utilizar grouby

```python
colunas = ['Atleta', 'Medalha']

medalhas_atletas = medalhas_Brasil.filter(items=colunas)

medalhas_atletas = medalhas_atletas.groupby('Atleta', as_index=False).count()

medalhas_atletas
```

##  Vamos transformar o tipo da coluna medalha de "object" para "int"


```python
medalhas_atletas['Medalha'] = medalhas_atletas['Medalha'].astype(int)

medalhas_atletas
```

```python
medalhas_atletas.info()
```
